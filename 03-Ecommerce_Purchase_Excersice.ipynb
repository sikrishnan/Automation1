1. Import pandas and read in the Ecommerce Purchases csv file and set it to a DataFrame called ecom.

```python
import pandas as pd
```

```python
ecom_data=pd.read_csv('Ecommerce Purchases',index_col=0)
ecom_data
```

```python
ecom=pd.DataFrame(ecom_data)
ecom
```

2. Check the head of the DataFrame.

```python
ecom_head=ecom.head()
ecom_head
```

3. How many rows and columns are there?

```python
ecom.info()
```

4. What is the average Purchase Price?

```python
avg_ecom=ecom['Purchase Price'].mean()
avg_ecom
```

5. What were the highest and lowest purchase prices?

```python
max_ecom = ecom['Purchase Price'].max()
max_ecom
```

```python
min_ecom = ecom['Purchase Price'].min()
min_ecom
```

6. How many people have English 'en' as their Language of choice on the website?

```python
english_users_count = (ecom['Language'] == 'en').sum()
english_users_count
```

7. How many people have the job title of "Lawyer" ?

```python
lawyer_count = (ecom['Job'] == 'Lawyer').sum()
lawyer_count
```

8. How many people made the purchase during the AM and how many people made the purchase during PM ?

```python
am_count = (ecom['AM or PM'] == 'AM').sum()
pm_count = (ecom['AM or PM'] == 'PM').sum()
print("Number of people purchases during AM:", am_count)
print("Number of people purchases during PM:", pm_count)
```

9. What are the 5 most common Job Titles?

```python
top_jobs = ecom['Job'].value_counts().head(5)
top_jobs
```

10. Someone made a purchase that came from Lot: "90 WT" , what was the Purchase Price for this transaction?

```python
purchase_price = ecom[ecom['Lot'] == '90 WT']['Purchase Price'].iloc[0]
purchase_price
```

11. What is the email of the person with the following Credit Card Number: 4926535242672853

```python
email = ecom.loc[ecom['Credit Card'] == 4926535242672853, 'Email'].iloc[0]
email
```

12. How many people have American Express as their Credit Card Provider and made a purchase above $95 ?

```python
american_express_above_95 = (ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)
count_people = american_express_above_95.sum()
count_people
```

13. Hard: How many people have a credit card that expires in 2025?

```python
expire_2025 = ecom['CC Exp Date'].str[-2:] == '25'
count_2025 = expire_2025.sum()
count_2025
```

14. Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...)

```python
ecom['Email Domain'] = ecom['Email'].str.split('@').str[1]
# print(ecom['Email Domain'])
top_email_providers = ecom['Email Domain'].value_counts().head(5)
top_email_providers
```
